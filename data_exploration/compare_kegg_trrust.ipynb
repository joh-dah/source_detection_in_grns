```python
import pandas as pd
df_gene_info = pd.read_csv('data/geneinfo_beta.txt', sep='\t')
L1000_genes = df_gene_info['gene_symbol'].tolist() 
L1000_ids = df_gene_info['gene_id'].tolist()
```

```python
# read kegg graph from ../data/topos/cleaned_kegg_cancer_graph.graphml
import networkx as nx
import matplotlib.pyplot as plt

G_kegg = nx.read_graphml('../data/topos/tgf_kegg.graphml')
# print the number of nodes and edges
print(f'Number of nodes: {G_kegg.number_of_nodes()}')
print(f'Number of edges: {G_kegg.number_of_edges()}')
# print the number of nodes that are in L1000 and KEGG
Kegg_in_l1000 = [node for node in G_kegg.nodes if node in L1000_genes]
print(f'Number of L1000 genes in KEGG: {len(Kegg_in_l1000)}')
# list the nodes that are in Kegg but not in L1000
Kegg_not_in_L1000 = [node for node in G_kegg.nodes if node not in L1000_genes]
print(f'Number of KEGG genes not in L1000: {len(Kegg_not_in_L1000)}')
# print the first 10 nodes that are in KEGG but not in L1000
print('KEGG genes not in L1000:', Kegg_not_in_L1000)
# strip the hsa: prefix from the Kegg_not_in_L1000 nodes
Kegg_not_in_L1000_stripped = [node.replace('hsa:', '') for node in Kegg_not_in_L1000]
print('KEGG genes not in L1000 (stripped):', Kegg_not_in_L1000_stripped)
# print how many of Kegg_not_in_L1000_stripped are in L1000_ids
Kegg_not_in_L1000_in_L1000_ids = [node for node in Kegg_not_in_L1000_stripped if node in L1000_ids]
print(f'Number of KEGG genes not in L1000 that are in L1000_ids: {len(Kegg_not_in_L1000_in_L1000_ids)}')
```

```python
# read kegg graph from ../data/topos/trrust.graphml
G_trrust = nx.read_graphml('../data/topos/trrust.graphml')
# print the number of nodes and edges
print(f'Number of nodes in TRRUST: {G_trrust.number_of_nodes()}')
print(f'Number of edges in TRRUST: {G_trrust.number_of_edges()}')
# print the number of nodes that are in L1000 and TRRUST
Trrust_in_l1000 = [node for node in G_trrust.nodes if node in L1000_genes]
print(f'Number of L1000 genes in TRRUST: {len(Trrust_in_l1000)}')
# list the nodes that are in TRRUST but not in L1000
Trrust_not_in_L1000 = [node for node in G_trrust.nodes if node not in L1000_genes]
print(f'Number of TRRUST genes not in L1000: {len(Trrust_not_in_L1000)}')
# print the first 10 nodes that are in TRRUST but not in L1000
print('TRRUST genes not in L1000:', Trrust_not_in_L1000)
```

```python
# reade the genes listed in ../data/mapk_nodes.txt
with open('../data/mapk_nodes.txt', 'r') as f:
    mapk_nodes = f.read().splitlines()
# print the number of nodes in mapk_nodes
print(f'Number of nodes in mapk_nodes: {len(mapk_nodes)}')
# print the amount of nodes in mapk_nodes that are also in trrust
mapk_nodes_in_trrust = [node for node in mapk_nodes if node in G_trrust.nodes]
print(f'Number of mapk_nodes in TRRUST: {len(mapk_nodes_in_trrust)}')
G_trrust_kegg_mapk = G_trrust.subgraph(mapk_nodes).copy()
# print the number of nodes and edges in the subgraph
print(f'Number of nodes in TRRUST subgraph: {G_trrust_kegg_mapk.number_of_nodes()}')
print(f'Number of edges in TRRUST subgraph: {G_trrust_kegg_mapk.number_of_edges()}')
# drop all isolated nodes from the subgraph
G_trrust_kegg_mapk.remove_nodes_from(list(nx.isolates(G_trrust_kegg_mapk)))
# print the number of nodes and edges in the subgraph after dropping isolated nodes
print(f'Number of nodes in TRRUST subgraph after dropping isolated nodes: {G_trrust_kegg_mapk.number_of_nodes()}')
print(f'Number of edges in TRRUST subgraph after dropping isolated nodes: {G_trrust_kegg_mapk.number_of_edges()}')
#print number and size of the weakly connected components
wccs = list(nx.weakly_connected_components(G_trrust_kegg_mapk))
print(f'Number of weakly connected components: {len(wccs)}')
print(f'Size of weakly connected components: {[len(wcc) for wcc in wccs]}')
# select the largest weakly connected component as graph
largest_wcc = max(wccs, key=len)
G_trrust_kegg_mapk_largest_wcc = G_trrust_kegg_mapk.subgraph(largest_wcc).copy()
# print the number of nodes and edges in the largest weakly connected component
print(f'Number of nodes in largest weakly connected component: {G_trrust_kegg_mapk_largest_wcc.number_of_nodes()}')
print(f'Number of edges in largest weakly connected component: {G_trrust_kegg_mapk_largest_wcc.number_of_edges()}')
# draw the largest weakly connected component
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G_trrust_kegg_mapk_largest_wcc, seed=42)
nx.draw(G_trrust_kegg_mapk_largest_wcc, pos, with_labels=True, node_size=50, font_size=8)
plt.title('Largest Weakly Connected Component of TRRUST-Kegg-MAPK Subgraph')
plt.show()
# print name of all nodes in the largest weakly connected component
print('Nodes in largest weakly connected component:', list(G_trrust_kegg_mapk_largest_wcc.nodes))
```

```python
# get the intersection of the two graphs
G_intersection = nx.intersection(G_kegg, G_trrust)
# print the number of nodes and edges in the intersection
print(f'Number of nodes in intersection: {G_intersection.number_of_nodes()}')
print(f'Number of edges in intersection: {G_intersection.number_of_edges()}')
# print number of nodes and edges from kegg that are not in the intersection
G_kegg_not_in_intersection = G_kegg.number_of_nodes() - G_intersection.number_of_nodes()
print(f'Number of nodes in KEGG not in intersection: {G_kegg_not_in_intersection}')
G_kegg_edges_not_in_intersection = G_kegg.number_of_edges() - G_intersection.number_of_edges()
print(f'Number of edges in KEGG not in intersection: {G_kegg_edges_not_in_intersection}')
# print number of nodes and edges from trrust that are not in the intersection
G_trrust_not_in_intersection = G_trrust.number_of_nodes() - G_intersection.number_of_nodes()
print(f'Number of nodes in TRRUST not in intersection: {G_trrust_not_in_intersection}')
G_trrust_edges_not_in_intersection = G_trrust.number_of_edges() - G_intersection.number_of_edges()
print(f'Number of edges in TRRUST not in intersection: {G_trrust_edges_not_in_intersection}')
# draw the intersection graph
plt.figure(figsize=(10, 10))
nx.draw(G_intersection, with_labels=True, node_size=50, font_size=8)
plt.title('Intersection of KEGG and TRRUST Graphs')
plt.show()
```

```python
# print number and size of components with more than one node
components = [c for c in nx.weakly_connected_components(G_intersection) if len(c) > 1]
print(f'Number of components with more than one node: {len(components)}')
# print the size of each component
for i, component in enumerate(components):
    print(f'Component {i+1} size: {len(component)}')
```

```python
# select and display the largest component
largest_component = max(components, key=len)
print(f'Largest component size: {len(largest_component)}')
# print the nodes in the largest component
print('Nodes in the largest component:', largest_component)
# draw the largest component
plt.figure(figsize=(10, 10))
# layout the graph using a spring layout
pos = nx.spring_layout(G_intersection.subgraph(largest_component))
nx.draw(G_intersection.subgraph(largest_component), pos, with_labels=True, node_size=50, font_size=8)
plt.title('Largest Component of Intersection Graph')
plt.show()
```

```python
# --- Build intersection graph manually considering interaction type ---
G_intersection = nx.DiGraph()

# Only add nodes that are shared
common_nodes = set(G_kegg.nodes()).intersection(G_trrust.nodes())
G_intersection.add_nodes_from(common_nodes)

# For edges: include only those that exist in both graphs with matching interaction
for u, v, d1 in G_kegg.edges(data=True):
    if G_trrust.has_edge(u, v):
        d2 = G_trrust.get_edge_data(u, v)
        print(f"{d1.get("interaction")} vs {d2.get("2")}")

        # Require matching 'interaction' attribute (can be adjusted if needed)
        if d1.get("interaction") == d2.get("2"):
            G_intersection.add_edge(u, v, interaction=d1["interaction"])

# --- Reporting ---
print(f' Number of nodes in intersection: {G_intersection.number_of_nodes()}')
print(f' Number of edges in intersection: {G_intersection.number_of_edges()}')

G_kegg_not_in_intersection = G_kegg.number_of_nodes() - G_intersection.number_of_nodes()
print(f' Nodes in KEGG not in intersection: {G_kegg_not_in_intersection}')

G_kegg_edges_not_in_intersection = G_kegg.number_of_edges() - G_intersection.number_of_edges()
print(f' Edges in KEGG not in intersection: {G_kegg_edges_not_in_intersection}')

G_trrust_not_in_intersection = G_trrust.number_of_nodes() - G_intersection.number_of_nodes()
print(f' Nodes in TRRUST not in intersection: {G_trrust_not_in_intersection}')

G_trrust_edges_not_in_intersection = G_trrust.number_of_edges() - G_intersection.number_of_edges()
print(f' Edges in TRRUST not in intersection: {G_trrust_edges_not_in_intersection}')

# --- Visualization ---
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G_intersection, seed=42)
labels = nx.get_node_attributes(G_intersection, "label")
edge_colors = [
    {"activation": "green", "inhibition": "red"}.get(d["interaction"], "gray")
    for _, _, d in G_intersection.edges(data=True)
]
nx.draw(
    G_intersection,
    pos,
    with_labels=True,
    labels=labels if labels else None,
    node_size=100,
    font_size=8,
    edge_color=edge_colors,
    arrows=True,
)
plt.title("Intersection of KEGG and TRRUST Graphs (Matching Edge Types)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
