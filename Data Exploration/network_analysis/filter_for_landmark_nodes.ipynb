```python
# create a list from landmark_genes.txt
landmark_genes = []
with open('../data/landmark_genes.txt') as f:
    for line in f:
        landmark_genes.append(line.strip())
```

```python
print(len(landmark_genes))
```

```python
import pandas as pd

df_network = pd.read_csv('../data/ACH-000012_HCC827.csv', index_col='Row')
df_network
```

```python
# create a new dataframe only with columns that are in landmark_genes
df_network_landmark = df_network[df_network.columns.intersection(landmark_genes)]
```

```python
df_network_landmark = df_network_landmark[df_network_landmark.index.isin(landmark_genes)]
df_network_landmark
```

```python
# get a list of row names
row_names = df_network_landmark.index.tolist()
print(len(row_names))
# get the difference between the landmark_genes and row_names
diff = list(set(landmark_genes) - set(row_names))
print(len(diff))
# add the missing genes to the dataframe. fill the values with 0
for gene in diff:
    df_network_landmark.loc[gene] = 0
```

```python
#sort columns and rows alphabetically
df_network_landmark = df_network_landmark.sort_index(axis=0)
df_network_landmark = df_network_landmark.sort_index(axis=1)
```

```python
# store the new dataframe in a new csv file
df_network_landmark.to_csv('../data/ACH-000012_HCC827_landmark.csv')
```
